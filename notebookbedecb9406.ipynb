```python
import pandas as pd
import os

# Define function to convert bounding box coordinates to YOLO format
def convert_to_yolo(image_width, image_height, x, y, width, height):
    x_center = x / image_width + (width / 2) / image_width
    y_center = y / image_height + (height / 2) / image_height
    box_width = width / image_width
    box_height = height / image_height
    return x_center, y_center, box_width, box_height

# Define function to read CSV files and convert annotations
def convert_annotations(input_folder, output_folder, image_width, image_height):
    for filename in os.listdir(input_folder):
        if filename.endswith('.csv'):
            df = pd.read_csv(os.path.join(input_folder, filename))
            # Calculate YOLO coordinates and append to DataFrame
            df['x_center'], df['y_center'], df['box_width'], df['box_height'] = convert_to_yolo(image_width, image_height, df['x'], df['y'], df['width'], df['height'])
            # Convert class labels to integers
            df['label'] = df['label'].astype(int)
            # Write DataFrame to CSV in YOLOv8 format
            df[['label', 'x_center', 'y_center', 'box_width', 'box_height']].to_csv(os.path.join(output_folder, os.path.splitext(filename)[0] + '.txt'), index=False, header=False, sep=' ')

# Example usage
input_folder = "/kaggle/working/content/bboxes"
output_folder = '/kaggle/working/content/labels'
image_width = 1280  # Width of the images in pixels
image_height = 1280  # Height of the images in pixels
convert_annotations(input_folder, output_folder, image_width, image_height)
```

```python
import os
import shutil
import random

def move_files(source_images_folder, source_labels_folder, destination_folder, num_files):
    # Create destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    try:
        # Get list of files in images folder
        image_files = os.listdir(source_images_folder)
        # Randomly select num_files number of files
        selected_image_files = random.sample(image_files, num_files)

        # Move selected image files to destination folder
        for file in selected_image_files:
            shutil.move(os.path.join(source_images_folder, file), destination_folder)

            # Move corresponding label files to destination folder
            label_file = os.path.splitext(file)[0] + '.txt'
            shutil.move(os.path.join(source_labels_folder, label_file), destination_folder)
    except FileNotFoundError as e:
        print(f"Error: {e}")

# Define paths
source_images_folder = '/kaggle/working/train/images'
source_labels_folder = '/kaggle/working/train/labels'
destination_folder = '/kaggle/working/val'
num_files = 80

# Move files
move_files(source_images_folder, source_labels_folder, destination_folder, num_files)
```

```python
import os
import shutil

def separate_files(input_folder, output_folder_txt, output_folder_png):
    # Create output folders if they don't exist
    os.makedirs(output_folder_txt, exist_ok=True)
    os.makedirs(output_folder_png, exist_ok=True)

    # Iterate through files in the input folder
    for filename in os.listdir(input_folder):
        # Check if the file is a .txt file
        if filename.endswith('.txt'):
            shutil.move(os.path.join(input_folder, filename), output_folder_txt)
        # Check if the file is a .png file
        elif filename.endswith('.png'):
            shutil.move(os.path.join(input_folder, filename), output_folder_png)
        # Skip files of other formats
        else:
            print(f"Ignoring file: {filename}")

# Define paths
input_folder = '/kaggle/working/val'
output_folder_txt = '/kaggle/working/val/labels'
output_folder_png = '/kaggle/working/val/images'

# Separate files
separate_files(input_folder, output_folder_txt, output_folder_png)
```
